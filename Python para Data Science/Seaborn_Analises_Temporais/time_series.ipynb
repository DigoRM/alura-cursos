```python
!pip install seaborn==0.9.0
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
ax = sns.lineplot(x='mes',y='vendas',data=alucar)
ax.figure.set_size_inches(12,6)
ax.set_title('Vendas Alucar de 2017 e 2018',loc='left',fontsize=20)
ax.set_xlabel('Tempo',fontsize=14)
ax.set_ylabel('Vendas',fontsize=14)
```

```python
# creating another column => Growth (diff())
alucar['growth'] = alucar['vendas'].diff()
alucar
```

```python
def plotar(titulo,labelx,labely,x,y,dataset):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x=x ,y=y,data=dataset)
    ax.figure.set_size_inches(12,6)
    ax.set_title(titulo,loc='left',fontsize=20)
    ax.set_xlabel(labelx,fontsize=14)
    ax.set_ylabel(labely,fontsize=14)
    
```

```python
plotar('Growth of Alucar Sales','Tempo','Growth','mes','growth',alucar)
```

```python
# Aceleration of sales growth
alucar['aceleration'] = alucar['growth'].diff()
alucar
```

```python
plotar('Aceleration of Alucar Sales','Tempo','Aceleration','mes','aceleration',alucar)
```

```python
# como plotar os 3 graficos juntos:
plt.figure(figsize=(16,12))
ax = plt.subplot(3,1,1)
ax.set_title('Alucar sales between 2017 e 2018', fontsize=18,loc='left')
sns.lineplot(x='mes', y='vendas', data=alucar)
plt.subplot(3,1,2)
sns.lineplot(x='mes', y='growth', data=alucar)
plt.subplot(3,1,3)
sns.lineplot(x='mes', y='aceleration', data=alucar)
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(alucar['vendas'])
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação das vendas', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar['vendas'])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação das crescimento', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar['growth'][1:])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação das aceleração', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar['aceleration'][2:])
ax = ax
```

# Assinantes Newsletter

```python
assinantes = pd.read_csv('newsletter_alucar.csv')
assinantes['mes'] = pd.to_datetime(assinantes['mes'])
assinantes
```

```python
assinantes['aumento'] = assinantes['assinantes'].diff()
assinantes['aceleracao'] = assinantes['aumento'].diff()
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
  plt.figure(figsize=(22,22))
  ax = plt.subplot(3,1,1)
  ax.set_title(titulo, fontsize=18,loc='left')
  sns.lineplot(x=x, y=y1, data=dataset)
  plt.subplot(3,1,2)
  sns.lineplot(x=x, y=y2, data=dataset)
  plt.subplot(3,1,3)
  sns.lineplot(x=x, y=y3, data=dataset)
  ax = ax
```

```python
plot_comparacao('mes', 'assinantes', 'aumento', 'aceleracao', 
                assinantes, 'Análise de assinantes da newsletter')
```

```python
# vendas diárias da Chocolura
vendas_por_dia = pd.read_csv('vendas_por_dia.csv')
vendas_por_dia['aumento'] = vendas_por_dia['vendas'].diff()
vendas_por_dia['aceleracao'] = vendas_por_dia['aumento'].diff()
vendas_por_dia
```

```python
plot_comparacao('dia', 'vendas', 'aumento', 'aceleracao',
               vendas_por_dia, 'Análise de vendas de Outubro e Novembro - Chocolura')
```

```python
vendas_por_dia['dia'] = pd.to_datetime(vendas_por_dia['dia'])
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.strftime("%A")
vendas_por_dia
```

```python
vendas_por_dia['dia_da_semana'].unique()
```

```python
traducao_dias = {'Monday':'Segunda', 'Tuesday':'Terça', 'Wednesday':'Quarta', 
                 'Thursday':'Quinta', 'Friday':'Sexta', 'Saturday':'Sábado',
                'Sunday':'Domingo'}
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(traducao_dias)
vendas_por_dia
```

```python
group_vendas = vendas_por_dia.groupby('dia_da_semana').mean().round()
group_vendas
```

```python
# Correlação das vendas diárias
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação das vendas diárias', fontsize=18, x=0.3, y=0.95)
autocorrelation_plot(vendas_por_dia['vendas'])
ax = ax
```

```python
# Correlação das cresimento
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação do aumento diário', fontsize=18, x=0.3, y=0.95)
autocorrelation_plot(vendas_por_dia['aumento'][1:])
ax = ax
```

```python
# Correlação da aceleração
ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação da acelereção diária', fontsize=18, x=0.3, y=0.95)
autocorrelation_plot(vendas_por_dia['aceleracao'][2:])
ax = ax
```

# StatsModel

```python
from statsmodels.tsa.seasonal import seasonal_decompose
chocolura['mes']=pd.to_datetime(chocolura['mes'])
```

```python
resultado = seasonal_decompose(chocolura['vendas'], period=3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ({
    'observacao':observacao,
    'tendencia':tendencia,
    'sazonalidade':sazonalidade,
    'ruido':ruido
})

resultado = pd.DataFrame(data)
resultado
```

```python
plot_comparacao(resultado.index, 'observacao', 'tendencia', 'sazonalidade', resultado,
               'Exemplo de Statsmodels')
```
